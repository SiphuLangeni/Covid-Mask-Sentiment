```python
from os import environ
from tweepy import OAuthHandler
from tweepy import Cursor
from tweepy import API
import pandas as pd
```

```python
# API keys from .env
consumer_key = environ.get('CONSUMER_KEY')
consumer_secret = environ.get('CONSUMER_SECRET')
access_token = environ.get('ACCESS_TOKEN')
access_token_secret = environ.get('ACCESS_TOKEN_SECRET')
```

```python
# Authentication
auth = OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)

api = API(auth_handler=auth,
          wait_on_rate_limit=True,
          wait_on_rate_limit_notify=True)
```

```python
query = '(covid19 OR covid-19 or coronavirus) AND mask -filter:retweets'
```

```python
# Create an empty dictionary of data to be collected for each tweet
tweet_data = {'date_created': [],
              'tweet_id': [],
              'name': [],
              'user_name': [],
              'user_location': [],
              'geo': [],
              'coordinates': [],
              'place': [],
              'tweet': [],
              'tweet_length': [],
              'hashtags': []
}


for tweet in Cursor(api.search, q=query, lang='en', tweet_mode='extended').items(5):
    tweet_data['date_created'].append(tweet.created_at)
    tweet_data['tweet_id'].append(tweet.id_str)
    tweet_data['name'].append(tweet.user.name)
    tweet_data['user_name'].append(tweet.user.screen_name)
    tweet_data['user_location'].append(tweet.user.location)
    tweet_data['geo'].append(tweet.geo)
    tweet_data['coordinates'].append(tweet.coordinates)
    tweet_data['place'].append(tweet.place)
    tweet_data['tweet'].append(tweet.full_text)
    tweet_data['tweet_length'].append(len(tweet.full_text))
    hashtags = []
    for hashtag in tweet.entities['hashtags']:
        hashtags.append(hashtag['text'])
    tweet_data['hashtags'].append(hashtags)
```

```python
# Create a dataframe from the tweet data
Tweets = pd.DataFrame(tweet_data)

Tweets
```
