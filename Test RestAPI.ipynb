```python
from os import environ
from tweepy import OAuthHandler
from tweepy import Cursor
from tweepy import API
```

```python
# API keys from .env
consumer_key = environ.get('CONSUMER_KEY')
consumer_secret = environ.get('CONSUMER_SECRET')
access_token = environ.get('ACCESS_TOKEN')
access_token_secret = environ.get('ACCESS_TOKEN_SECRET')
```

```python
# Authentication
auth = OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)

api = API(auth_handler=auth,
          wait_on_rate_limit=True,
          wait_on_rate_limit_notify=True)
```

```python
# Query variable to include covid and mask keywords, exclude retweets
query = '(covid19 OR covid-19 or coronavirus or pandemic) AND mask -filter:retweets'
```

```python
# Print the full text of n number of tweets
for i, tweet in enumerate(Cursor(api.search, q=query, lang='en', tweet_mode='extended').items(100)):
    print(i, tweet.full_text)
```
